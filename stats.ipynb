```python
import pandas as pd
```

```python
paths=["evals/CRW_seed42_2026-04-01_05-27-20_std/ppo_ep.csv", 
       "evals/CRW_seed99_2026-04-01_11-10-58_m3/ppo_ep.csv",
       "evals/CRW_seed99_2026-04-01_11-33-03_m1/ppo_ep.csv",
       "evals/CRW_seed42_2026-04-01_07-01-43_swl/ppo_ep.csv",
       "evals/CRW_seed99_2026-04-01_07-06-31_jackals/ppo_ep.csv",
       "evals/CRW_seed99_2026-04-01_08-16-09_sac/sac_ep.csv",
       "evals/CRW_seed99_2026-04-01_09-12-38_sac_jackals/sac_ep.csv",
       "evals/EE_seed42_2026-04-01_06-05-21_std/ppo_ep.csv",
       "evals/EE_seed42_2026-04-01_08-14-32_sac/sac_ep.csv",
       "evals/EE_seed99_2026-04-01_09-20-04_sac_swl/sac_ep.csv",
       "evals/LPOI_seed42_2026-04-01_09-07-22_sac_pigeons/sac_ep.csv",
       "evals/LPOI_seed99_2026-04-01_06-33-53_std/ppo_ep.csv",
       "evals/LPOI_seed99_2026-04-01_08-57-03_sac/sac_ep.csv",
       "evals/POI_seed42_2026-04-01_06-34-33_std/ppo_ep.csv",
       "evals/POI_seed42_2026-04-01_08-54-44_sac/sac_ep.csv",
       "evals/POI_seed99_2026-04-01_07-11-28_pigeons/ppo_ep.csv",
       "evals/CRW_seed42_2026-04-01_05-27-20_std/CentroidStandoff_ep.csv",
       "evals/EE_seed42_2026-04-01_06-05-21_std/CentroidStandoff_ep.csv",
       "evals/POI_seed42_2026-04-01_06-34-33_std/CentroidStandoff_ep.csv",
       "evals/LPOI_seed99_2026-04-01_06-33-53_std/CentroidStandoff_ep.csv",
       "evals/POI_seed99_2026-04-01_07-11-28_pigeons/CentroidStandoff_ep.csv",
       "evals/CRW_seed99_2026-04-01_07-06-31_jackals/CentroidStandoff_ep.csv",
       "evals/CRW_seed42_2026-04-01_07-01-43_swl/CentroidStandoff_ep.csv"
]
```

```python
results = []

cols = ["reward", "r_monitoring", "p_disturbance"]

for path in paths:
    df = pd.read_csv(path)

    stats = df[cols].agg(["mean", "std"]).T

    results.append({
        "file": path.split("/")[-2],
        "reward_mean": stats.loc["reward", "mean"],
        "reward_std": stats.loc["reward", "std"],
        "monitor_mean": stats.loc["r_monitoring", "mean"],
        "monitor_std": stats.loc["r_monitoring", "std"],
        "dist_mean": stats.loc["p_disturbance", "mean"],
        "dist_std": stats.loc["p_disturbance", "std"],
    })

summary_df = pd.DataFrame(results)

print(summary_df)
```

```python
summary_df["reward"] = summary_df.apply(
    lambda r: f"{r.reward_mean:.3f} ± {r.reward_std:.3f}", axis=1
)

summary_df["r_monitoring"] = summary_df.apply(
    lambda r: f"{r.monitor_mean:.3f} ± {r.monitor_std:.3f}", axis=1
)

summary_df["p_disturbance"] = summary_df.apply(
    lambda r: f"{r.dist_mean:.3f} ± {r.dist_std:.3f}", axis=1
)

final_table = summary_df[["file", "reward", "r_monitoring", "p_disturbance"]]

print(final_table)
```
